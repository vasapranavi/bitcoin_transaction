```python
# This is single customer transaction data
def single_customer_data(ano):
    # Creating index for the data
    ano['index1'] = ano.index
    
    # Initialising final dataframe
    result = pd.DataFrame(columns = ['customer_id', 'created_at', 'type', 'currency', 'transaction_name', 'transaction_amount'])
    
    # Taking only XBT transactions
    xbt = ano[ano['currency'] == 'XBT']
    
    for i in range(0, len(ano)):
        # Taking XBT tranctions previous to current line
        temp = xbt[xbt['index1']<ano.iloc[i][8]]
        
        # For transaction type 1
        if i == 0:
            if ano.iloc[i][1] == 1 and ano.iloc[i][6]=='XBT':
                to_add = pd.DataFrame({'customer_id': ano.iloc[i][5], 'created_at': ano.iloc[i][2], 'type': ano.iloc[i][1], 'currency': ano.iloc[i][6], 'transaction_name': 'ReceiveBTC','transaction_amount': ano.iloc[i][4]}, index=[0])
                result = result.append(to_add, ignore_index=True)
            elif ano.iloc[i][1] == 1 and ano.iloc[i][6]!='XBT':
                to_add = pd.DataFrame({'customer_id': ano.iloc[i][5], 'created_at': ano.iloc[i][2], 'type': ano.iloc[i][1], 'currency': ano.iloc[i][6], 'transaction_name': 'FiatDeposit','transaction_amount': ano.iloc[i][4]}, index=[0])
                result = result.append(to_add, ignore_index=True)
        
        # For transaction 1 preceding with transaction 2
        elif ano.iloc[i-1][1] == 2 and ano.iloc[i][1] == 1:
            if (ano.iloc[i][3] - ano.iloc[i-1][3]) < 0:
                to_add = pd.DataFrame({'customer_id': ano.iloc[i][5], 'created_at': ano.iloc[i][2], 'type': ano.iloc[i][1], 'currency': ano.iloc[i][6], 'transaction_name': 'SendBTC','transaction_amount': (ano.iloc[i][3] - ano.iloc[i-1][3])}, index=[0])
            else:
                to_add = pd.DataFrame({'customer_id': ano.iloc[i][5], 'created_at': ano.iloc[i][2], 'type': ano.iloc[i][1], 'currency': ano.iloc[i][6], 'transaction_name': 'ReceiveBTC','transaction_amount': (ano.iloc[i][3] - ano.iloc[i-1][3])}, index=[0])
            result = result.append(to_add, ignore_index=True)

        elif ano.iloc[i-1][1] != 2 and ano.iloc[i][1] == 1 and i!=0 and len(temp)!=0:
            to_add = pd.DataFrame({'customer_id': ano.iloc[i][5], 'created_at': ano.iloc[i][2], 'type': ano.iloc[i][1], 'currency': ano.iloc[i][6], 'transaction_name': 'ReceiveBTC','transaction_amount': ano.iloc[i][4] - temp.iloc[len(temp)- 1][4]}, index=[0])
            result = result.append(to_add, ignore_index=True)
        
        # For transaction 4 and 'XBT'
        elif (ano.iloc[i][1] == 4)  and (ano.iloc[i][6] == 'XBT'):
            to_add = pd.DataFrame({'customer_id': ano.iloc[i][5], 'created_at': ano.iloc[i][2], 'type': ano.iloc[i][1], 'currency': ano.iloc[i][6], 'transaction_name': 'Bid','transaction_amount': 0}, index=[0])
            result = result.append(to_add, ignore_index=True)
            
        # For transaction 4 and not 'XBT'
        elif (ano.iloc[i][1] == 4)  and (ano.iloc[i][6] != 'XBT'):
            to_add = pd.DataFrame({'customer_id': ano.iloc[i][5], 'created_at': ano.iloc[i][2], 'type': ano.iloc[i][1], 'currency': ano.iloc[i][6], 'transaction_name': 'Ask','transaction_amount': 0}, index=[0])
            result = result.append(to_add, ignore_index=True)
            
        # For transaction 14
        elif (ano.iloc[i][1] == 14) and (ano.iloc[i][6] == 'XBT'):
            to_add = pd.DataFrame({'customer_id': ano.iloc[i][5], 'created_at': ano.iloc[i][2], 'type': ano.iloc[i][1], 'currency': ano.iloc[i][6], 'transaction_name': 'SellBTC','transaction_amount': ano.iloc[i][4] - temp.iloc[len(temp)- 1][4]}, index=[0])
            result = result.append(to_add, ignore_index=True)
            
        # For tansaction 15
        elif (ano.iloc[i][1] == 15) and (ano.iloc[i][6] == 'XBT') and len(temp)!=0:
            to_add = pd.DataFrame({'customer_id': ano.iloc[i][5], 'created_at': ano.iloc[i][2], 'type': ano.iloc[i][1], 'currency': ano.iloc[i][6], 'transaction_name': 'BuyBTC','transaction_amount': ano.iloc[i][4] - temp.iloc[len(temp)- 1][4]}, index=[0])
            result = result.append(to_add, ignore_index=True)
            
        # For transaction 7
        elif (ano.iloc[i][1] == 7) and (ano.iloc[i][6] == 'XBT') and len(temp)!=0:
            
            to_add = pd.DataFrame({'customer_id': ano.iloc[i][5], 'created_at': ano.iloc[i][2], 'type': ano.iloc[i][1], 'currency': ano.iloc[i][6], 'transaction_name': 'TradeBuy','transaction_amount': ano.iloc[i][4] - temp.iloc[len(temp)- 1][4]}, index=[0])
            result = result.append(to_add, ignore_index=True)
    return result
    
```

```python
# Checking with given customer data
result = single_customer_data(ano)
result
```

```python
customers = ledger_account_t.customer_id.unique()
customers
```

```python
# Generating for all the customers
task_1 = pd.DataFrame(columns = ['customer_id', 'created_at', 'type', 'currency', 'transaction_name', 'transaction_amount'])
for customer in customers:
    customer_transaction = single_customer_data(ledger_account_t[ledger_account_t['customer_id']==customer])
    task_1 = task_1.append(customer_transaction, ignore_index=True)
    
# Final Output for task 1
task_1
```
